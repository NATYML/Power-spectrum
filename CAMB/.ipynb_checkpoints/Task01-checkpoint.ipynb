```python
%matplotlib inline
import sys, platform, os
import os
from matplotlib import pyplot as plt
import numpy as np
print('Using CAMB installed at '+ os.path.realpath(os.path.join(os.getcwd(),'..')))
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'..')))
import camb
from camb import model, initialpower
from scipy import integrate
```

```python
    #Constants 
h = 0.675

ombh2 = 0.022
omch2 = 0.122
om0h2 = omch2 + ombh2
ome0 = 0.3089
omel = 0.6911
Tcmb = 2.725
ns=0.965

#Definitions for the master function T(k)
zeq = 2.5e4*om0h2*(Tcmb/2.7)**(-4)
a1 = (46.9*om0h2)**0.67*(1+(32.1*om0h2)**(-0.532))
a2 = (12*om0h2)**0.424*(1+(45*om0h2)**(-0.582))
alpha = a1**(-ombh2/om0h2)*a2**(-(ombh2/om0h2)**3)
```

```python
#Function that only depends on sc
def trans_func(k):
    q = k*(Tcmb/2.7)**2/(om0h2*alpha**0.5*(1-ombh2/om0h2)**0.6)
    return np.log(1+2.34*q)*(1+13*q+(10.5*q)**2+(10.4*q)**3+(6.51*q)**4)**(-1/4.)/(2.34*q)

#Power spectrum (PS) that includes the initial PS and dependency on k of the transfer function
def P_init(k):
    return k**ns*trans_func(k)**2

def g2(z):
    return ome0*(1+z)**3+(1-ome0-omel)*(1+z)**2+omel

#Growth function, dependency on z of the transfer function 
def gr_func(z):
    omez = ome0*(1+z)**3*g2(z)**(-1)
    omelz = omel*g2(z)**(-1)
    return (1+zeq)/(1+z)*5*omez*0.5*( omez**(4/7.)-omelz+(1+0.5*omez)*(1+omelz/70.) )**(-1)

#Power spectrum evolved using the trasnfer function
def P(z,k):
    return P_init(k)*(gr_func(z)/gr_func(0))**2

#Window function 
def Wrk(R,k):
    return 3*(np.sin(k*R)-k*R*np.cos(k*R))/((k*R)**2.)    

#Normalization using sigma 8 as shown in report
def Norm(R,z):
    return 1./(integrate.quad(lambda k:P(z,k)*(Wrk(R,k)*k)**2,0.,np.inf)[0]/(2*np.pi**2*(1+z)**(-2)))
```

```python
#Different redshifts to obtain the power spectrum using the analytical treatment
z = np.array([0.,1.,2.,3,10.,50.,100.])
k = np.linspace(1e-4,1e0,100)
plt.figure(figsize=(14,14))
plt.subplot(2,2,1)
for i in xrange(4,7):
    plt.loglog(k,Norm(8.,z[i])*P(z[i],k),label="z=%d"%z[i])
plt.title('Matter power at z= %s with analytical expression'%z[4:7])   
plt.xlabel('k/h Mpc');
plt.legend()
plt.subplot(2,2,2)
for i in xrange(0,4):
    plt.loglog(k,Norm(8.,z[i])*P(z[i],k),label="z=%d"%z[i])
plt.title('Matter power at z= %s with analytical expression'%z[0:5]) 
plt.xlabel("k")
plt.legend()
```

```python
#Different redshifts to obtain the power spectrum using CAMB

#Now get matter power spectra and sigma8 at redshift 0 and 0.8
pars = camb.CAMBparams()
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122)
pars.set_dark_energy() #re-set defaults
pars.InitPower.set_params(ns=0.965)
#Not non-linear corrections couples to smaller scales than you want
pars.set_matter_power(redshifts=[0.,1.,2.,3,10.,50.,100.], kmax=2.0)

#Linear spectra
pars.NonLinear = model.NonLinear_none
results = camb.get_results(pars)
kh, z, pk = results.get_matter_power_spectrum(minkh=1e-4, maxkh=1, npoints = 200)
s8 = np.array(results.get_sigma8())
```

```python
plt.figure(figsize=(14,14))
plt.subplot(2,2,1)
for i in xrange(4,7):
    plt.loglog(kh, pk[i,:],label="z=%d"%z[i])
plt.title('Matter power at z= %s with CAMB'%z[4:7])   
plt.xlabel('k/h Mpc');
plt.legend()
plt.subplot(2,2,2)
for i in xrange(0,4):
    plt.loglog(kh, pk[i,:],label="z=%d"%z[i])
plt.title('Matter power at z= %s with CAMB'%z[0:5]) 
plt.xlabel("k")
plt.legend()
```
