```python
import numpy as np
import matplotlib.pylab as plt
%pylab inline
```

```python
Pk_box = np.loadtxt("../Interpolation/Pk_Box1000x1024_z_0_NGRID_1024_CIC.dat.Whead")
Pk_TSC_1e11 = np.loadtxt( "../../output_files/Pk_halosTSC_1e11_256.dat" )
Pk_TSC_1e12 = np.loadtxt( "../../output_files/Pk_halosTSC_1e12_128.dat" )
Pk_TSC_1e13 = np.loadtxt( "../../output_files/Pk_halosTSC_1e13_64.dat" )
```

```python
Pk_CIC_1e11 = np.loadtxt( "../../output_files/PS_CIC_1e11.dat" )
Pk_CIC_1e12 = np.loadtxt( "../../output_files/PS_CIC_1e12.dat" )
Pk_CIC_1e13 = np.loadtxt( "../../output_files/PS_CIC_1e13.dat" )
```

```python
plt.figure(figsize=(7,7))
plt.loglog(Pk_box[:,0],Pk_box[:,1], label="CIC for box")
plt.loglog( Pk_TSC_1e11[:,0],Pk_TSC_1e11[:,1], "." )
plt.loglog( Pk_TSC_1e12[:,0],Pk_TSC_1e12[:,1], "." )
plt.loglog( Pk_TSC_1e13[:,0],Pk_TSC_1e13[:,1], "." )
plt.title("P(k) for TSC scheme and different halo populations ")
plt.grid()
plt.legend(loc="lower left")
```

```python
plt.figure(figsize=(6,6))
plt.loglog( Pk_TSC_1e11[:,0],Pk_TSC_1e11[:,1], ".", label= "TSC halo pop 1e11" )
plt.loglog( Pk_CIC_1e11[:,0],Pk_CIC_1e11[:,1],".", label="CIC halo pop 1e11" )
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(6,6))
plt.loglog( Pk_TSC_1e12[:,0],Pk_TSC_1e12[:,1], ".", label="TSC halo pop 1e12" )
plt.loglog( Pk_CIC_1e12[:,0],Pk_CIC_1e12[:,1],".", label="CIC halo pop 1e12" )
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(6,6))
plt.loglog( Pk_TSC_1e13[:,0],Pk_TSC_1e13[:,1], ".", label="TSC halo pop 1e13" )
plt.loglog( Pk_CIC_1e13[:,0],Pk_CIC_1e13[:,1],".", label="CIC halo pop 1e13" )
plt.grid()
plt.legend()
```
