```python
%matplotlib inline
import sys, platform, os
import os
from matplotlib import pyplot as plt
import numpy as np
print('Using CAMB installed at '+ os.path.realpath(os.path.join(os.getcwd(),'..')))
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'..')))
import camb
from camb import model, initialpower
from scipy import integrate
from scipy.interpolate import interp1d  
```

```python
#Now get matter power spectra and sigma8 at redshift 0 and 0.8
pars = camb.CAMBparams()
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122)
pars.set_dark_energy() #re-set defaults
pars.InitPower.set_params(ns=0.965)
```

```python
                        #Constant of Box 400 

#Constants 
h = 0.675

ombh2 = 0.022
omch2 = 0.122
om0h2 = omch2 + ombh2
ome0 = 0.3089
omel = 0.6911
Tcmb = 2.725
ns = 0.965

#Definitions for the master function T(k)
zeq = 2.5e4*om0h2*(Tcmb/2.7)**(-4)
a1 = (46.9*om0h2)**0.67*(1+(32.1*om0h2)**(-0.532))
a2 = (12*om0h2)**0.424*(1+(45*om0h2)**(-0.582))
alpha = a1**(-ombh2/om0h2)*a2**(-(ombh2/om0h2)**3)
```

```python
                            # Theoretical functions 

#Function that only depends on sc
def trans_func(k):
    q = k*(Tcmb/2.7)**2/(om0h2*alpha**0.5*(1-ombh2/om0h2)**0.6)
    return np.log(1+2.34*q)*(1+13*q+(10.5*q)**2+(10.4*q)**3+(6.51*q)**4)**(-1/4.)/(2.34*q)

#Power spectrum (PS) that includes the initial PS and dependency on k of the transfer function
def P_init(k):
    return k**ns*trans_func(k)**2

def g2(z):
    return ome0*(1+z)**3+(1-ome0-omel)*(1+z)**2+omel

#Growth function, dependency on z of the transfer function 
def gr_func(z):
    omez = ome0*(1+z)**3*g2(z)**(-1)
    omelz = omel*g2(z)**(-1)
    return (1+zeq)/(1+z)*5*omez*0.5*( omez**(4/7.)-omelz+(1+0.5*omez)*(1+omelz/70.) )**(-1)

#Power spectrum evolved using the trasnfer function
def P(z,k):
    return P_init(k)*(gr_func(z)/gr_func(0))**2

#Window function 
def Wrk(R,k):
    return 3*( np.sin(k*R)-k*R*np.cos(k*R) )/((k*R)**2.)

#Normalization using sigma 8 as shown in report
def Norm(R,z):
    return 1./(integrate.quad(lambda k:P(z,k)*(Wrk(R,k)*k)**2,0.,np.inf)[0]/(2*np.pi**2*(1+z)**(-2)))
```

```python
                        # Theoretical functions and interpolation function for CAMB 
                     # Bias and Sigma can be obtained with CAMB (interp) or theoretical approach
                                        # interp = "True"

def interp( z0 ):

    KMAX = 1000
    #Not non-linear corrections couples to smaller scales than you want
    pars.set_matter_power( redshifts=[z0], kmax=KMAX )

    #Linear spectra
    pars.NonLinear = model.NonLinear_none
    results = camb.get_results(pars)
    kh, z, pk = results.get_matter_power_spectrum( minkh=1e-4, maxkh=KMAX, npoints = 500 )
    
    #Non-Linear spectra (Halofit)
    pars.NonLinear = model.NonLinear_both
    results.calc_power_spectra(pars)
    kh_nonlin, z_nonlin, pk_nonlin = results.get_matter_power_spectrum(minkh=1e-4, maxkh=KMAX, npoints = 500)    
    
    #Linear spectra in z interpolated
    Pl = interp1d(kh, pk[0,:],kind='linear')
    Pnl = interp1d( kh_nonlin, pk_nonlin[0,:],kind='linear')
    
    return Pl, Pnl


#Calculating sigma(R)
def sigma( R,z0, Pl, NORM=1.0, interp=False, plot = False  ):
    
    if plot == True:
        
        KMAX = 100
        plt.figure(figsize=(5,5))    
        a1 = np.linspace(1e-3,KMAX-1,500)
        y = Pl(a1)*(Wrk(R,a1)*a1)**2
        plt.plot(a1,y,".")
        plt.xlim(0,5)
        plt.grid()
    
    if interp == "True":
            # CAMB
                rs = NORM*integrate.quad( lambda k:Pl(k)*(Wrk(R,k)*k)**2,1e-4,10, points= [1e-4])[0]/(2*np.pi**2*(1+z0)**(-2) )
                return np.sqrt(rs)
    else: 
            #Theoretical result
            rs = NORM*integrate.quad( lambda k:P(z0,k)*(Wrk(R,k)*k)**2,0.,np.inf)[0]/(2*np.pi**2*(1+z0)**(-2) )
            return np.sqrt(rs)

def bias( R, z, Pl, interp=False, NORM=1.0,plot= False ):
    
    #Critical density
    delta_c = 1.686
    #Peak height
    nu = delta_c/sigma(R,z, Pl, NORM, interp, plot)
    delta = 200.
    y = np.log10( delta )
        
    A = 1. + 0.24*y*np.exp(-(4./y)**4)
    a = 0.44*y - 0.88
    B = 0.183
    b = 1.5
    C = 0.019 + 0.107*y + 0.19*np.exp(-(4./y)**4)
    c = 2.4
    
    return delta_c/nu, 1 - A*nu**a/(nu**a+delta_c**a) + B*nu**b + C*nu**c
    
```

```python
                            # Adjust sigma 8 from Klypin (2010) https://arxiv.org/pdf/1002.3660v1.pdf


def sigma_appr( M ):
    
    y =  ( M/1.e12 )**(-1.)
       
    return ( 16.9*y**0.41/( 1+1.102*y**0.2+6.22*y**0.333) )


# Bias from adjusted sigma
def bias_appr( M ):
    
    #Critical density
    delta_c = 1.686
    #Peak height
    nu = delta_c/sigma_appr( M )
    delta = 200.
    y = np.log10( delta )
        
    A = 1. + 0.24*y*np.exp(-(4./y)**4)
    a = 0.44*y - 0.88
    B = 0.183
    b = 1.5
    C = 0.019 + 0.107*y + 0.19*np.exp(-(4./y)**4)
    c = 2.4
    
    return delta_c/nu,1 - A*nu**a/(nu**a+delta_c**a) + B*nu**b + C*nu**c
    
```

```python
                # Normalization for theoretical result and CAMB 

z = 0
NORM = Norm(8.,z)
```

```python
                        # From Mass to Radio dependence 

Msun = 1.9891e30#kg
kmtops = 3.086e13 # km to parsec
# kg/m^3
rho_c = 1e-26*(1./Msun)*( (1000/1.)*(kmtops)*(1e6/1.) )**3 #Msun/Mpc
#Mean density of the Universe rhom = omega0 rho_c
om0 = om0h2/h**2
rhom = rho_c*om0


delta_c = 200.
Np = 2000
sigma0 = np.zeros( Np )
sigma_approx = np.zeros( Np )
bias0 = np.zeros( Np )

M = np.linspace( 1e8,1e16,Np  )# Msun

R = (3*M/(4*np.pi*rhom))**(1/3.)
```

```python
CAMB = "False"
#CAMB = "True"

if CAMB == "True": 
    
    # Interpolation function if CAMB is used
    Pl = interp( z )
    
else:
    
    Pl = np.sin


for i in xrange(len(R)):
    
    sigma0[i],bias0[i] = bias( R[i],z,Pl,CAMB,NORM )

# Peak hight 
#nu = delta_c/sigma0    
```

```python
#--------------------------------------------------------------------------------

b1e11 = np.sqrt(Pkhal_1e11[:,1]/Pkhal_1e11_box[:,1])
LEN = len(b1e11)
Pkhal_1e11 = np.concatenate( (Pkhal_1e11, np.reshape(b1e11,(LEN,1))), axis=1 )

#--------------------------------------------------------------------------------

b1e12 = np.sqrt(Pkhal_1e12[:,1]/Pkhal_1e12_box[:,1])
LEN = len(b1e12)
Pkhal_1e12 = np.concatenate( (Pkhal_1e12, np.reshape(b1e12,(LEN,1))), axis=1 )


#--------------------------------------------------------------------------------

b1e13 = np.sqrt(Pkhal_1e13[:,1]/Pkhal_1e13_box[:,1])
LEN = len(b1e13)
Pkhal_1e13 = np.concatenate( (Pkhal_1e13, np.reshape(b1e13,(LEN,1))), axis=1 )
```

```python

#--------------------------------------------------------------------------------
                            # Halo Population 1e11
    
    
deltak = 0.03
kmin = Pkhal_1e11[0,0]
kmax = Pkhal_1e11[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e11[:,0])-min(Pkhal_1e11[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e11 = np.zeros((3,1))

for i in xrange( ITER - 2 ):
    
    temp = Pkhal_1e11[Pkhal_1e11[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e11 = np.append( b_adj_1e11,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e11)    
b_adj_1e11 = np.reshape( b_adj_1e11[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e11[:,0], b_adj_1e11[:,2], deg = 1)
bias_lineal_1e11 = np.poly1d(z)

#--------------------------------------------------------------------------------
                            # Halo Population 1e12
deltak = 0.01

kmin = Pkhal_1e12[0,0]
kmax = Pkhal_1e12[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e12[:,0])-min(Pkhal_1e12[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e12 = np.zeros((3,1))

for i in xrange( ITER ):
    
    temp = Pkhal_1e12[Pkhal_1e12[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e12 = np.append( b_adj_1e12,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e12)    
b_adj_1e12 = np.reshape( b_adj_1e12[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e12[:,0], b_adj_1e12[:,2], deg = 1)
bias_lineal_1e12 = np.poly1d(z)


#--------------------------------------------------------------------------------
                            # Halo Population 1e13

deltak = 0.01

kmin = Pkhal_1e13[0,0]
kmax = Pkhal_1e13[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e13[:,0])-min(Pkhal_1e13[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e13 = np.zeros((3,1))

for i in xrange( ITER ):
    
    temp = Pkhal_1e13[Pkhal_1e13[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e13 = np.append( b_adj_1e13,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e13)    
b_adj_1e13 = np.reshape( b_adj_1e13[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e13[:,0], b_adj_1e13[:,2], deg = 1)
bias_lineal_1e13 = np.poly1d(z)

```
