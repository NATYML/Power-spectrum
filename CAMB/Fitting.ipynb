```python
#Now get matter power spectra and sigma8 at redshift 0 and 0.8
pars = camb.CAMBparams()
pars.set_cosmology(H0=h*100, ombh2 = ombh2, omch2 = omch2 )
pars.set_dark_energy() #re-set defaults
pars.InitPower.set_params( ns = ns )
```

```python
                        # Theoretical functions and interpolation function for CAMB 
                     # Bias and Sigma can be obtained with CAMB (interp) or theoretical approach
                                        # interp = "True"

def interp( z0 ):

    KMAX = 1000
    #Not non-linear corrections couples to smaller scales than you want
    pars.set_matter_power( redshifts=[z0], kmax=KMAX )

    #Linear spectra
    pars.NonLinear = model.NonLinear_none
    results = camb.get_results(pars)
    kh, z, pk = results.get_matter_power_spectrum( minkh=1e-4, maxkh=KMAX, npoints = 500 )
    
    #Non-Linear spectra (Halofit)
    pars.NonLinear = model.NonLinear_both
    results.calc_power_spectra(pars)
    kh_nonlin, z_nonlin, pk_nonlin = results.get_matter_power_spectrum(minkh=1e-4, maxkh=KMAX, npoints = 500)    
    s8 = np.array(results.get_sigma8())
    
    #Linear spectra in z interpolated
    Pl = interp1d(kh, pk[0,:],kind='linear')
    Pnl = interp1d( kh_nonlin, pk_nonlin[0,:],kind='linear')
    
    return Pl, Pnl, s8


#Calculating sigma(R)
def sigma( R,z0, Pl, NORM=1.0, interp=False, plot = False  ):
    
    if plot == True:
        
        KMAX = 100
        plt.figure(figsize=(5,5))    
        a1 = np.linspace(1e-3,KMAX-1,500)
        y = Pl(a1)*(Wrk(R,a1)*a1)**2
        plt.plot(a1,y,".")
        plt.xlim(0,5)
        plt.grid()
    
    if interp == "True":
            # CAMB
                rs = NORM*integrate.quad( lambda k:Pl(k)*(Wrk(R,k)*k)**2,1e-4,10, points= [1e-4])[0]/(2*np.pi**2*(1+z0)**(-2) )
                return np.sqrt(rs)
    else: 
            #Theoretical result
            rs = NORM*integrate.quad( lambda k:P(z0,k)*(Wrk(R,k)*k)**2,0.,np.inf)[0]/(2*np.pi**2*(1+z0)**(-2) )
            return np.sqrt(rs)

def bias( R, z, Pl, interp=False, NORM=1.0,plot= False ):
    
    #Critical density
    delta_c = 1.686
    #Peak height
    nu = delta_c/sigma(R,z, Pl, NORM, interp, plot)
    delta = 200.
    y = np.log10( delta )
        
    A = 1. + 0.24*y*np.exp(-(4./y)**4)
    a = 0.44*y - 0.88
    B = 0.183
    b = 1.5
    C = 0.019 + 0.107*y + 0.19*np.exp(-(4./y)**4)
    c = 2.4
    
    return delta_c/nu, 1 - A*nu**a/(nu**a+delta_c**a) + B*nu**b + C*nu**c
```

```python

#--------------------------------------------------------------------------------
                            # Halo Population 1e11
    
    
deltak = 0.03
kmin = Pkhal_1e11[0,0]
kmax = Pkhal_1e11[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e11[:,0])-min(Pkhal_1e11[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e11 = np.zeros((3,1))

for i in xrange( ITER - 2 ):
    
    temp = Pkhal_1e11[Pkhal_1e11[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e11 = np.append( b_adj_1e11,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e11)    
b_adj_1e11 = np.reshape( b_adj_1e11[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e11[:,0], b_adj_1e11[:,2], deg = 1)
bias_lineal_1e11 = np.poly1d(z)

#--------------------------------------------------------------------------------
                            # Halo Population 1e12
deltak = 0.01

kmin = Pkhal_1e12[0,0]
kmax = Pkhal_1e12[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e12[:,0])-min(Pkhal_1e12[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e12 = np.zeros((3,1))

for i in xrange( ITER ):
    
    temp = Pkhal_1e12[Pkhal_1e12[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e12 = np.append( b_adj_1e12,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e12)    
b_adj_1e12 = np.reshape( b_adj_1e12[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e12[:,0], b_adj_1e12[:,2], deg = 1)
bias_lineal_1e12 = np.poly1d(z)


#--------------------------------------------------------------------------------
                            # Halo Population 1e13

deltak = 0.01

kmin = Pkhal_1e13[0,0]
kmax = Pkhal_1e13[0,0] + deltak

ITER =  int( np.ceil( ( max(Pkhal_1e13[:,0])-min(Pkhal_1e13[:,0]) )/deltak ) )
INTV = 0.8

b_adj_1e13 = np.zeros((3,1))

for i in xrange( ITER ):
    
    temp = Pkhal_1e13[Pkhal_1e13[:,0]>=kmin]
    
    temp = temp[temp[:,0] <= kmax]
    mean = np.mean( temp[:,2] )
    sig = np.std( temp[:,2] )
    low = mean - INTV*sig
    upp = mean + INTV*sig
    temp1 = temp[temp[:,2] >= low]
    temp1 = temp1[temp1[:,2] <= upp]
    b_adj_1e13 = np.append( b_adj_1e13,temp1 )
    kmin = kmax
    kmax = kmin + deltak

L1 = len(b_adj_1e13)    
b_adj_1e13 = np.reshape( b_adj_1e13[3:L1],( (L1-3)/3,3 ))

z = np.polyfit( b_adj_1e13[:,0], b_adj_1e13[:,2], deg = 1)
bias_lineal_1e13 = np.poly1d(z)

```
